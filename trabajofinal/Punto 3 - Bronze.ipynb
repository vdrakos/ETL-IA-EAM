```python
from pyspark.sql import functions as F
from pyspark.sql import SparkSession
```

```python
CATALOG = "insurance_claims"
BRONZE_SCHEMA = "bronze"
SILVER_SCHEMA = "silver"
GOLD_SCHEMA = "gold"
LANDING_SCHEMA = "landing"

VOLUME_NAME = "raw_files"


CUSTOMERS_PATH = f"/Volumes/{CATALOG}/{BRONZE_SCHEMA}/{VOLUME_NAME}/customers/customers.csv"
POLICIES_PATH = f"/Volumes/{CATALOG}/{BRONZE_SCHEMA}/{VOLUME_NAME}/policies/policies.csv"
CLAIMS_PATH = f"/Volumes/{CATALOG}/{BRONZE_SCHEMA}/{VOLUME_NAME}/claims/claims.csv"
METADATA_IMG_PATH = f"/Volumes/{CATALOG}/{BRONZE_SCHEMA}/{VOLUME_NAME}/claim_images_metadata/image_metadata.csv"
TELEMATICS_PATH = f"/Volumes/{CATALOG}/{BRONZE_SCHEMA}/{VOLUME_NAME}/telematics"
IMAGE_PATH = f"/Volumes/{CATALOG}/{BRONZE_SCHEMA}/{VOLUME_NAME}/claim_imgs/"
TRAINING_IMAGE_PATH = f"/Volumes/{CATALOG}/{BRONZE_SCHEMA}/{VOLUME_NAME}/training_imgs/"

spark = SparkSession.builder.getOrCreate()
```

```python
#Leemos customers
customers_bz = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(CUSTOMERS_PATH)
)

#Leemos policies
policies_bz = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(POLICIES_PATH)
)

#Leemos claims
claims_bz = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(CLAIMS_PATH)
)

#Leemos metadata imgs
metadata_bz = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(METADATA_IMG_PATH)
)

```

```python
#Leemos telematics
telematics_bz = spark.read.parquet(TELEMATICS_PATH)
display(telematics_bz)
```

```python
#Leemos imgs
claims_imgs_bz = spark.read.format("binaryFile").load(IMAGE_PATH)
claims_imgs_bz.count()
```

```python
#Leemos training imgs
training_imgs_bz = spark.read.format("binaryFile").load(TRAINING_IMAGE_PATH)
training_imgs_bz.count()
```

```python
# Guardar tablas bronze

customers_bz.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", True) \
    .saveAsTable(f"{CATALOG}.{BRONZE_SCHEMA}.customers_bz") 

claims_bz.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", True) \
    .saveAsTable(f"{CATALOG}.{BRONZE_SCHEMA}.claims_bz") 

policies_bz.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", True) \
    .saveAsTable(f"{CATALOG}.{BRONZE_SCHEMA}.policies_bz") 

metadata_bz.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", True) \
    .saveAsTable(f"{CATALOG}.{BRONZE_SCHEMA}.metadata_bz") 

telematics_bz.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", True) \
    .saveAsTable(f"{CATALOG}.{BRONZE_SCHEMA}.telematics_bz") 

claims_imgs_bz.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", True) \
    .saveAsTable(f"{CATALOG}.{BRONZE_SCHEMA}.claims_imgs_bz")


```

```python
training_imgs_bz.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", True) \
    .saveAsTable(f"{CATALOG}.{BRONZE_SCHEMA}.training_imgs_bz")
```
